```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from util import *
from nilearn import plotting
import matplotlib.pyplot as plt

rsfc_la, rsfc_ra, rsfc_lh, rsfc_rh = load_rsfc_data()

rsfc_la_corr = np.zeros((116, 116))
rsfc_ra_corr = np.zeros((116, 116))
rsfc_lh_corr = np.zeros((116, 116))
rsfc_rh_corr = np.zeros((116, 116))
rsfc_la_mean = np.mean(rsfc_la, axis=0)
rsfc_ra_mean = np.mean(rsfc_ra, axis=0)
rsfc_lh_mean = np.mean(rsfc_lh, axis=0)
rsfc_rh_mean = np.mean(rsfc_rh, axis=0)

for i in range(116):
    
    rsfc_la_corr[40, i] = rsfc_la_mean[i]
    rsfc_la_corr[i, 40] = rsfc_la_mean[i]
    
    rsfc_ra_corr[41, i] = rsfc_ra_mean[i]
    rsfc_ra_corr[i, 41] = rsfc_ra_mean[i]
    
    rsfc_lh_corr[36, i] = rsfc_lh_mean[i]
    rsfc_lh_corr[i, 36] = rsfc_lh_mean[i]
    
    rsfc_rh_corr[37, i] = rsfc_rh_mean[i]
    rsfc_rh_corr[i, 37] = rsfc_rh_mean[i]

coordinates = plotting.find_parcellation_cut_coords(labels_img='/Users/lzn/Downloads/BrainNetViewer_20191031/Data/ExampleFiles/AAL90/aal.nii') 
```

```python
plotting.plot_connectome(adjacency_matrix=rsfc_la_corr, 
                         node_coords=coordinates,
                         node_color='auto',
                         node_size=15,
                         edge_cmap=plt.cm.YlOrRd,
                         edge_vmin=0,
                         edge_vmax=0.7636029156666316,
                         display_mode='ortho',
                         annotate=False,
                         alpha=0.7,
                         colorbar=False)

#plt.savefig('./rsfc_la.pdf', dpi=666)
```

```python
plotting.plot_connectome(adjacency_matrix=rsfc_lh_corr, 
                         node_coords=coordinates,
                         node_color='auto',
                         node_size=15,
                         edge_cmap=plt.cm.PuBu,
                         edge_vmin=0,
                         edge_vmax=0.7636029156666316,
                         display_mode='ortho',
                         annotate=False,
                         alpha=0.7,
                         colorbar=False)

#plt.savefig('./rsfc_lh.pdf', dpi=666)
```
