```python
df_ss = DataFrame(
    {
        'MMS similarity': mms_1,
        'rs-FC similarity': rsfc_1,
        'IMQ similarity': fits_1,
        'rs-FC similarity (quartile)': rsFC_q_1,
    })

df_so = DataFrame(
    {
        'MMS similarity': mms_2,
        'rs-FC similarity': rsfc_2,
        'IMQ similarity': fits_2,
        'rs-FC similarity (quartile)': rsFC_q_2,
    })

df_os_la = DataFrame(
    {
        'MMS similarity': mms_3,
        'rs-FC similarity': rsfc_3,
        'IMQ similarity': fits_3,
        'rs-FC similarity (quartile)': rsFC_q_3,
    })
```

# Figure 7(b)

```python
fig = plt.figure(figsize=(10, 6.18))

ax = fig.gca(projection='3d')

y_surf = np.arange(df_ss['MMS similarity'].min(),
                   df_ss['MMS similarity'].max(), 0.1)
x_surf = np.arange(df_ss['rs-FC similarity'].min(),
                   df_ss['rs-FC similarity'].max(), 0.1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)
z_surf = model_1.coefs['Estimate'][
    0] + model_1.coefs['Estimate'][1] * y_surf + model_1.coefs['Estimate'][
        2] * x_surf + model_1.coefs['Estimate'][3] * x_surf * y_surf
surf = ax.plot_surface(x_surf, 
                       y_surf, 
                       z_surf,
                       rstride=1,
                       cstride=1,
                       cmap=cm.viridis,
                       vmin=-1.4808488665379376,
                       vmax=0.7020550108228392,
                       linewidth=0)

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

#ax.set_ylabel('MMS similarity')
#ax.set_xlabel('rs-FC similarity')
#ax.set_zlabel('SS similarity')
#ax.set_xticks([2, -3,], minor=False)
ax.invert_xaxis()
ax.tick_params(axis='x',labelsize=14)
ax.tick_params(axis='y',labelsize=14)
ax.tick_params(axis='z',labelsize=14)
ax.set_facecolor('white')
plt.grid(True)

#plt.savefig('./dra_cb.pdf', dpi=666)
```
